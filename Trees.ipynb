```python
#PostOrder Implementaion:(Recursive) -> lrR
def postorder_traversal_recursive(root):
    result = []
    def Traverse(node):
        if not node:
            return
        
        Traverse(node.left)
        Traverse(node.right)
        result.append(node.val)
    Traverse(root)
    return results

#PostOrder Implementaion:(Iterative Approach) -> lrR
def postorder_traversal_recursive(root):
    if not root:
        return
    stack = [root]
    results = []
    while stack:
        node = stack.pop()
        results.append(node.val)

        if node.left:
            results.append(node.left)
        if node.right:
            results.append(node.right)

    return results[::-1]
```

```python
#Inorder Implementaion:(Recursive) -> lRr
def inorder_traversal_recursive(root):
    result = []
    def Traverse(node):
        if not node:
            return
        
        Traverse(node.left)
        result.append(node.val)
        Traverse(node.right)

    Traverse(root)
    return results

#InOrder Implementaion:(Iterative Approach) -> lRr
def inorder_traversal_recursive(root):
    stack = []
    results = []
    current = root
    while stack or current:
        while stack:
            stack.append(current)
            current = current.left
        
        current = stack.pop()
        results.append(current.val)
        current = current.right

    return results
```

```python
#Level Order
from collections import deque

def level_order_traversal(root):
    if not root:
        return []
    queue = deque([root])
    results = []

    while queue:
        level = []
        for i in range(len(queue)):
            node = queue.popleft()
            level.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        results.append(level)

    return results
```

```python
#Unifeied Traverdal
def unified_traversal(root):
    if not root:
        return [], [], []
    stack = [(root,0)]
    preorder = []
    inorder = []
    postorder = []
    while stack:
        node, state = stack.pop()
        if state == 0:
            preorder.append(node.val)
            stack.append((node,1))
            if node.left:
                stack.append((node.left,0))
        
        if state == 1:
            inorder.append(node.val)
            stack.append((node,2))
            if node.right:
                stack.append((node.right,0))
        
        if state == 2:
            postorder.append(node.val)

    return inorder,postorder,preorder


```

```python
#Invert Binary Tree
def invertTree(self, root):
    if not root:
        return 
    if not root.left or not root.right:
        return root
    
    temp = root.left
    root.right = root.left
    root.left = temp
    self.invertTree(root)
```

```python
#104. Maximum Depth of Binary Tree
from collections import deque
class Solution(object):
    def maxDepth(self, root):
        if not root:
            return 0 
        queue = deque([root])
        result = []
        depth = 0
        while queue:
            for i in range(len(queue)):
                node = queue.popleft()
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            depth += 1
        return depth
        
#TC - O(N)
#SC - O(N)
```

```python
#110. Balanced Binary Tree
class Solution(object):
    def isBalanced(self, root):
        def height(node):
            if not node:    
                return 0
            
            left_height = height(node.left)
            right_height = height(node.right)

            if left_height == -1 or right_height == -1:
                return -1
            
            if(abs(left_height-right_height)>1):
                return -1

            return max(left_height,right_height) + 1
        
        return height(root) != -1
#TC - O(N)
#SC - O(H)
```

```python
#Diamter of Binary Tree
class Solution(object):
    def diameterOfBinaryTree(self, root):
        self.maxdiameter = 0
        def diameter(node):

            if not node:
                return 0
            
            lh = diameter(node.left)
            rh = diameter(node.right)
            self.maxdiameter = max(self.maxdiameter,lh+rh)

            return max(lh,rh) + 1
        
        diameter(root)
        return self.maxdiameter
#TC - O(N)
#SC - O(H)
```

```python
#100. Same Tree
class Solution(object):
    def isSameTree(self, p, q):
        # Base cases
        if not p and not q:  # Both nodes are None
            return True
        if not p or not q:  # One is None, the other is not
            return False
        if p.val != q.val:  # Node values differ
            return False
        
        # Recursive check for left and right subtrees
        return self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)
#TC O(n)
#SC O(h)
```

```python
#101. Symmetric Tree
class Solution(object):
    def isSymmetric(self, root):
        def isMirror(left, right):
            # Base cases
            if not left and not right:
                return True
            if not left or not right:
                return False
            if left.val != right.val:
                return False
            
            # Recursive check for symmetry
            return isMirror(left.left, right.right) and isMirror(left.right, right.left)
        
        # Start the recursion with the left and right subtrees
        return isMirror(root.left, root.right) if root else True
#TC - O(N)
#SC - O(H)
```

```python
#637. Average of Levels in Binary Tree
from collections import deque
class Solution(object):
    def averageOfLevels(self, root):
        if not root:
            return []
        results = []
        queue = deque([root])
        while queue:
            level_sum = 0
            level_count = len(queue)
            for _ in range(level_count):
                node = queue.popleft()
                level_sum += node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                
            results.append(level_sum/float(level_count))
        return results
#TC - O(N) - each node once
#SC - O(N) - queue       
```

```python
#222. Count Complete Tree Nodes
class Solution(object):
    def countNodes(self, root):
        def height(node,goLeft):
            height = 0
            while node:
                height += 1
                node = node.left if goLeft else node.right
            return height
        if not root:
            return 0
        
        left_height = height(root,True)
        right_height = height(root,False)

        if left_height == right_height:
            #call 2^h - 1
            return (1 << left_height) - 1
        else:
            return self.countNodes(root.left) + self.countNodes(root.right) + 1
#TC - O(logN) - for computing the height
#    - O(logN) - for Recursive calls
# -> O(logN * logN)
#SC = O(logN)
```

```python
#112. Path Sum
class Solution(object):
    def hasPathSum(self, root, targetSum):
        if not root:
            return False
        if not root.left and not root.right:
            return root.val == targetSum
        
        rem_sum = targetSum - root.val
        return self.hasPathSum(root.left,rem_sum) or self.hasPathSum(root.right,rem_sum)
#TC - O(N) - each node once
#SC - O(H) 
```
