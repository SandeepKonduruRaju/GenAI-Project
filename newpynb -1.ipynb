```python
class Stack(object):

    def __init__(self):  # Corrected from _init_ to __init__
        self.items = []

    def isEmpty(self):
        return len(self.items) == 0  # More explicit check for emptiness
    
    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.isEmpty():  # Added check to avoid popping from an empty stack
            return self.items.pop()
        else:
            raise IndexError("pop from empty stack")  # Raise error if stack is empty
    
    def peek(self):
        if not self.isEmpty():  # Added check to avoid peeking into an empty stack
            return self.items[-1]  # More Pythonic way to get the last item
        else:
            raise IndexError("peek from empty stack")  # Raise error if stack is empty
    
    def size(self):
        return len(self.items)
```

```python
# Create a new stack
stack = Stack()

# Check if the stack is empty
print(stack.isEmpty())  # Output: True

# Push items onto the stack
stack.push(1)
stack.push(2)
stack.push(3)

# Check the size of the stack
print(stack.size())  # Output: 3

# Peek at the top item
print(stack.peek())  # Output: 3

# Pop an item
print(stack.pop())   # Output: 3

# Check the size again
print(stack.size())  # Output: 2
```

```python
#Implemantation of Queue
class Queue(object):

    def __init__(self):  # Corrected from _init_ to __init__
        self.items = []

    def isEmpty(self):
        return len(self.items) == 0  # More explicit check for emptiness
    
    def enqueue(self, item):
        self.items.insert(0,item)

    def dequeue(self):
        if not self.isEmpty():  # Added check to avoid popping from an empty stack
            return self.items.pop()
        else:
            raise IndexError("pop from empty stack")  # Raise error if stack is empty
        
    def size(self):
        return len(self.items)
```

```python
# Create a new queue
queue = Queue()

# Check if the queue is empty
print(queue.isEmpty())  # Output: True

# Enqueue items into the queue
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

# Check the size of the queue
print(queue.size())  # Output: 3

# Dequeue an item
print(queue.dequeue())  # Output: 1

# Check the size again
print(queue.size())  # Output: 2

# Peek at the next item (not implemented in this version)
```

```python
from collections import deque

class Deque:
    def __init__(self):
        self.items = deque()  # Initialize the deque

    def is_empty(self):
        return len(self.items) == 0  # Check if the deque is empty

    def add_front(self, item):
        self.items.appendleft(item)  # Add an item to the front of the deque

    def add_rear(self, item):
        self.items.append(item)  # Add an item to the rear of the deque

    def remove_front(self):
        if not self.is_empty():  # Check if the deque is not empty
            return self.items.popleft()  # Remove and return the front item
        else:
            raise IndexError("Cannot remove from an empty deque.")  # Raise error if empty

    def remove_rear(self):
        if not self.is_empty():  # Check if the deque is not empty
            return self.items.pop()  # Remove and return the rear item
        else:
            raise IndexError("Cannot remove from an empty deque.")  # Raise error if empty

    def size(self):
        return len(self.items)  # Return the number of items in the deque

    def peek_front(self):
        if not self.is_empty():  # Check if the deque is not empty
            return self.items[0]  # Return the front item without removing it
        else:
            raise IndexError("Cannot peek from an empty deque.")  # Raise error if empty

    def peek_rear(self):
        if not self.is_empty():  # Check if the deque is not empty
            return self.items[-1]  # Return the rear item without removing it
        else:
            raise IndexError("Cannot peek from an empty deque.")  # Raise error if empty
```

```python
# Create a new Deque instance
deque = Deque()

# Add items to both ends
deque.add_rear(1)
deque.add_rear(2)
deque.add_front(0)

# Display current items in deque
print("Current Deque:", list(deque.items))  # Output: [0, 1, 2]

# Remove items from both ends
print("Removed from front:", deque.remove_front())  # Output: 0
print("Removed from rear:", deque.remove_rear())    # Output: 2

# Current size of deque
print("Current size:", deque.size())                # Output: 1

# Peek at remaining items
print("Front item:", deque.peek_front())            # Output: 1
print("Rear item:", deque.peek_rear())              # Output: 1

# Final state of deque
print("Final Deque:", list(deque.items))            # Output: [1]
```

```python
#Balanced Parentheses Check
def is_balanced(parentheses):
    # Dictionary to hold matching pairs
    matching_parentheses = {')': '(', '}': '{', ']': '['}
    stack = []

    for char in parentheses:
        if char in matching_parentheses.values():  # If it's an opening bracket
            stack.append(char)  # Push it onto the stack
        elif char in matching_parentheses.keys():  # If it's a closing bracket
            if stack and stack[-1] == matching_parentheses[char]:  # Check for match
                stack.pop()  # Pop the matching opening bracket
            else:
                return False  # Mismatch found or stack is empty

    return len(stack) == 0  # If stack is empty, parentheses are balanced
```

```python
# Example usage
test_strings = [
    "()",        # Balanced
    "(())",      # Balanced
    "(()())",    # Balanced
    "({[]})",    # Balanced
    "((",        # Not balanced
    ")",         # Not balanced
    "([)]",      # Not balanced
    "{[()]}",    # Balanced
]

for s in test_strings:
    print(f"{s}: {is_balanced(s)}")
```

```python
class QueueUsingStacks:
    def __init__(self):
        self.stack1 = []  # Stack for enqueue operations
        self.stack2 = []  # Stack for dequeue operations

    def enqueue(self, item):
        # Push the item onto stack1
        self.stack1.append(item)

    def dequeue(self):
        # If both stacks are empty, raise an error
        if not self.stack1 and not self.stack2:
            raise IndexError("dequeue from empty queue")

        # If stack2 is empty, transfer elements from stack1 to stack2
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())

        # Pop the top item from stack2 and return it
        return self.stack2.pop()

    def is_empty(self):
        # The queue is empty if both stacks are empty
        return not self.stack1 and not self.stack2

    def size(self):
        # Return the total number of items in the queue
        return len(self.stack1) + len(self.stack2)

# Example usage
if __name__ == "__main__":
    queue = QueueUsingStacks()
    
    # Enqueue items
    queue.enqueue(1)
    queue.enqueue(2)
    queue.enqueue(3)

    print("Dequeue:", queue.dequeue())  # Output: 1
    print("Dequeue:", queue.dequeue())  # Output: 2

    queue.enqueue(4)
    
    print("Dequeue:", queue.dequeue())  # Output: 3
    print("Dequeue:", queue.dequeue())  # Output: 4

    print("Is empty?", queue.is_empty())  # Output: True
```
