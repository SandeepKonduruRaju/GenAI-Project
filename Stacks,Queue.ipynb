```python
#20. Valid Parentheses
class Solution(object):
    def isValid(self, s):
        #Create dict
        matching_parentheis = {')':'(',']':'[','}':'{'}
        stack = []

        for char in s:
            if char in matching_parentheis.values():
                stack.append(char)
            elif char in matching_parentheis.keys():
                if not stack or stack[-1] != matching_parentheis[char]:
                    return False
                stack.pop()
            else:
                return False
            
        return not stack
#Time - O(n) 
#algorithm iterates through each character in the input string exactly once.
#Operations like appending to or popping from the stack take  O(1) time.
#Thus, the time complexity is proportional to the size of the string.

#Space  - O(n)
#The stack is used to store unmatched opening brackets.
#In the worst case, when the input string consists of all opening brackets (e.g., "[[[{{("), the stack can grow to the size of the input.
#The dictionary for matching parentheses takes O(1) space because it is constant in size.
```

```python
class Solution(object):
    def nextGreaterElement(self, nums1, nums2):
        # Dictionary to store the next greater element for nums2
        next_greater = {}
        # Monotonic stack to keep track of elements in nums2
        stack = []

        # Traverse nums2 in reverse to find the next greater elements
        for num in reversed(nums2):
            # Maintain a decreasing order in the stack
            while stack and stack[-1] <= num:
                stack.pop()
            # If stack is not empty, the top element is the next greater
            next_greater[num] = stack[-1] if stack else -1
            # Push the current element onto the stack
            stack.append(num)

        # Map the results for nums1 using the next_greater dictionary
        return [next_greater[num] for num in nums1]
    
##TIme = O(n+ m)
#Space = O(n)
```
