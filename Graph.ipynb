```python
#BFS Traversal
from collections import deque
class solution:
    def bfs_of_graph(self,V,adj):
        visited = [False] * V
        bfs = []
        queue = deque([0])
        visited[0] = True

        while queue:
            node = queue.popleft()
            bfs.append(node)

            for neigh in adj[node]:
                if not visited[neigh]:
                    queue.append(neigh)
                    visited[neigh] = True
        return bfs
#TC - O(V+E) Adj List O(V^2) - Adj List
#SC - O(V+E) Adj List O(V^2) - Adj List
```

```python
#994. Rotting Oranges
from collections import deque
class Solution(object):
    def orangesRotting(self, grid):
        #intialization
        rows,colms = len(grid),len(grid[0])
        fresh,min = 0,0
        queue = deque()
        direction = [(0,1),(1,0),(-1,0),(0,-1)]

        for r in range(rows):
            for c in range(colms):
                if grid[r][c] == 2:
                    queue.append((r,c,0))
                elif grid[r][c] == 1:
                    fresh += 1
        
        while queue:
            x,y,min = queue.popleft()

            for dx,dy in direction:
                cx,cy = x+dx, y+dy

                if 0<=cx<rows and 0<=cy<colms and grid[cx][cy] == 1:
                    grid[cx][cy] = 2
                    queue.append((cx,cy,min+1))
                    fresh -= 1


        return min if fresh == 0 else -1
#TC - O(R*C)
#SC - O(R*C)
```

```python
#Implementation of graphs using BFS
from typing import List
from collections import deque

class Solution:
    # Function to detect cycle in an undirected graph using BFS
    def isCycle(self, V: int, adj: List[List[int]]) -> bool:
        visited = [False] * V  # To keep track of visited nodes
        
        # Helper function for BFS
        def bfs(start):
            queue = deque([(start, -1)])  # (node, parent)
            visited[start] = True  # Mark the starting node as visited
            
            while queue:
                node, parent = queue.popleft()  # Dequeue a node
                
                # Explore all the neighbors of the current node
                for neighbour in adj[node]:
                    # If the neighbor hasn't been visited, enqueue it and mark it
                    if not visited[neighbour]:
                        visited[neighbour] = True
                        queue.append((neighbour, node))  # Enqueue with the current node as parent
                    # If the neighbor is visited and is not the parent, a cycle is found
                    elif neighbour != parent:
                        return True
            return False
        
        # Check for cycles in each connected component
        for i in range(V):
            if not visited[i]:  # If not visited, start BFS
                if bfs(i):  # If BFS detects a cycle, return True
                    return True
        
        return False

#TC - O(V) - for loop and O(2E) - TWICE node is touched -O(V+E)
#sc - Visited Array: An array of size 𝑂(𝑉) Queue:𝑂(𝑉) Adjacency List: O(V+E) = O(V)
```

```python
#Implementation of graphs using DFS
from typing import List

class Solution:
    # Function to detect cycle in an undirected graph using DFS
    def isCycle(self, V: int, adj: List[List[int]]) -> bool:
        visited = [False] * V  # To keep track of visited nodes
        
        # Helper function for DFS
        def dfs(start, parent):
            visited[start] = True  # Mark the current node as visited
            
            # Explore all the neighbors of the current node
            for neighbour in adj[start]:
                # If the neighbour hasn't been visited, continue DFS from the neighbour
                if not visited[neighbour]:
                    if dfs(neighbour, start):  # Recursively visit the neighbour
                        return True
                # If the neighbour is visited and is not the parent, a cycle is found
                elif neighbour != parent:
                    return True
            
            return False
        
        # Check for cycles in each connected component
        for i in range(V):
            if not visited[i]:  # If not visited, start DFS
                if dfs(i, -1):  # -1 as the parent for the root node
                    return True
        
        return False
#TC - O(V+E)
#SC - O(V)
```
