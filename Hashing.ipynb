```python
#Two Sum
class Solution(object):
    def twoSum(self, nums, target):
        seen = {}
        for i in range(len(nums)):
            rem  = target - nums[i]
            if rem in seen:
                return i,seen[rem]
            else:
                seen[nums[i]] = i
# Time- O(n)
# Space - O(1)
```

```python
#Anagram
class Solution(object):
    def isAnagram(self, s, t):
        seen = {}
        for i in s:
            if i in seen:
                seen[i] += 1
            else:
                seen[i] = 1

        for j in t:
            if j in seen:
                seen[j] -= 1
            else:
                return False
        
        count = 0
        for count in seen.values():
            if count != 0:
                return False
            
        return True
#Time - O(n)
#Space - O(k) -  O(k), where k is the number of unique characters in s and t (since we store their frequencies in the seen dictionary).
```

```python
class Solution(object):
    def containsDuplicate(self, nums):
        seen = {}
        for i in nums:
            if i in seen:
                return True
            else:
                seen[i] = 1
            
        return False
#Time - O(n)
#Space-O(n)
```

```python
#349. Intersection of Two Arrays
class Solution(object):
    def intersection(self, nums1, nums2):
        a_set = set(nums1)
        result = []
        for i in nums2:
            if i in a_set and i not in result:
                result.append(i)
        
        return result

#Time - O( m + n * k )
#Space- O(min(m,n)

#Most optimized use 2 sets
class Solution(object):
    def intersection(self, nums1, nums2):
        a_set = set(nums1)
        b_set = set()
        result = []
        for i in nums2:
            if i in a_set:
                b_set.add(i)
          
        return list(b_set)
    
#Time - O( m + n )
#Space- O(n + min(n, m)) this is because the intersection can have at most min(n, m) elements.     
```

```python
#136. Single Number
class Solution(object):
    def singleNumber(self, nums):
        seen = {}
        for i in nums:
            if i in seen:
                seen[i] += 1
            else:
                seen[i] = 1
        count = 0
        for key,count in seen.items():
            if count !=  2:
                return key
#Time - O(n)
#Space-O(n)
```

```python
#Naive String Matching Algorithm
def naive_string_matching(text, pattern):
    n = len(text)
    m = len(pattern)
    
    results = []
    for i in range(0,n-m+1):
        ismatch = True
        for j in range(0,m):
            if text[i+j] != pattern[j]:
                ismatch = False
                break
        
        if ismatch:
            results.append(i)
    
    return results

#Time = O( n * m)
#Space - O(k)
```

```python
class SolutionofLong1(object):
    def lengthOfLongestSubstring(self, s):
        charset = set()
        l = 0
        res = 0
        for r in range(l,len(s)):
            while s[r] in charset:
                charset.remove(s[l])
                l += 1
            charset.add(s[r])
            res = max(res,r-l+1)
        return res
```

```python
solution = SolutionofLong1()
print(solution.lengthOfLongestSubstring("pwwkew"))  # Output: 3 ("abc")
```
